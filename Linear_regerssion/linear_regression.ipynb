```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
points = np.genfromtxt('linear_data.csv', delimiter=',')
X=[]
Y=[]
for i in range(len(points)):
    X.append(points[i][0])
    Y.append(points[i][1])
alpha=0.00001 #learning rate
```

```python
def compute_error(y_intercept,slope,X,Y):
    err=0
    for i in range(len(X)):
        err+=(y_intercept+slope*X[i]-Y[i])**2
    return (err/float(len(X)))/2
```

```python
def get_gradient_descent(X,Y,alpha):
    m=len(Y) #no_of training examples
    para=[0,0] #parameters
    gr0=0
    gr1=0
    num_iterations=5000
    
    while (num_iterations):
        num_iterations-=1
        for i in range(m):
            gr0+=para[0]+para[1]*X[i]-Y[i]
            gr1+=(para[0]+para[1]*X[i]-Y[i])*X[i]

        temp_para_0=para[0]-(alpha/m)*gr0
        temp_para_1=para[1]-(alpha/m)*gr1

        para[0]=temp_para_0
        para[1]=temp_para_1
    return para
```

```python
#print(get_gradient_decent(X,Y,alpha))
para=get_gradient_descent(X,Y,alpha)
print(para[0])
print(para[1])
x=np.linspace(20,80,500)
def myfunc(x):
    return x*para[1]+para[0]
my_y=list(map(myfunc,x))
plt.plot(x,my_y)
plt.scatter(X,Y)
plt.show()

print("Error initial")
print(compute_error(0,0,X,Y)) 

print("\nerror final")
print(compute_error(para[0],para[1],X,Y))
```
